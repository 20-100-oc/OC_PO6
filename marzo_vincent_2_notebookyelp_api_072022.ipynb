```python
# No more than 50 businesses can be fetched at a time, 
# so fetch multiple times
nb_max = 50

# coordinates of some french cities
coordinates = {'paris': [2.35023, 48.6459], 
               'toulouse': [1.44327, 43.60275], 
               'marseille': [5.37669, 43.29803], 
               'lyon': [4.83625, 45.75465], 
               'nantes': [-1.55148, 47.20584]
              }

business_ids = []
for lng, lat in coordinates.values():
    response = yelp_api.search_query(categories='restaurants', 
                                     longitude=lng, 
                                     latitude=lat, 
                                     limit=nb_max)
    for business in response['businesses']:
        business_ids.append(business['id'])
print('business_ids:', len(business_ids), '\n')

nb_reviews = 0
df_list = []
for i, business_id in enumerate(business_ids):
    response = yelp_api.reviews_query(id=business_id)
    df = pd.json_normalize(response, record_path=['reviews'])
    df_list.append(df)

df_final = pd.concat(df_list, axis=0, ignore_index=True)
df_final    
```
